```python
import seaborn as sns
sns.set_palette('husl')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
Quantity=df.groupby(['Product']).agg(quantity=('Quantity_Ordered','sum'))
sns.barplot(x='Product',y='quantity',data=Quantity)
plt.xlabel('PRODUCT')
plt.ylabel('QUANTITY')
plt.title('MOST SOLD PRODUCT')
plt.xticks(rotation='vertical')
plt.yticks(Quantity['quantity'])
plt.grid()
plt.show()
```

In this graph we can understand that AAA Batteries (4-pack) is most sold product 

# WHICH CITY HAS HIGHEST SALES?

```python
pd.pivot_table(data=df,index='City',values='Sales',aggfunc='sum')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
city_sales=pd.pivot_table(data=df,index='City',values='Sales',aggfunc='sum')
sns.barplot(x='City',y='Sales',data=city_sales)
sns.set_palette='husl'
plt.title('Highest sales')
plt.xlabel('City')
plt.ylabel('Sales')
plt.grid()
plt.xticks(rotation='vertical')
plt.yticks(city_sales['Sales'])
plt.show()
```

In this graph we can understand that San Francisco has highest sales and best city for sales

# CHANGE THE DATATYPE OF ORDAER_DATE TO DATETIME FORMAT

```python
df['Order_Date']=pd.to_datetime(df['Order_Date'])
df.dtypes
```

# TOTAL SALES OF PRODUCT IN DIFFERRENT YEARS

```python
pd.pivot_table(
    df,
    index=['Product',df['Order_Date'].dt.year],
    columns='City',
    values='Sales',
    aggfunc=sum).tail(10)
```

# ADD A NEW COLUMN CALLED DAY

```python
df['Day']=df['Order_Date'].dt.day_name()
df
```

# BEST DAY FOR SALES

```python
df.groupby(['Day']).agg( sales=('Sales','sum')).sort_values(by='sales',ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
day_sales=df.groupby(['Day']).agg( sales=('Sales','sum'))
sns.lineplot(x='Day',y='sales',data=day_sales)
plt.title('best day for sale')
plt.xlabel('day')
plt.ylabel('sales')
plt.grid()
plt.yticks(day_sales['sales'])
plt.show()
```

# ADD COLUMN TOTAL SALES PER CUSTOMER

```python
df['Total_sales_per_customer']=df['Sales'].groupby(df['Order_ID']).transform('sum')
df
```

# ADD COLUMN TOTAL ORDER PER CUSTOMER

```python
df['Total_order_per_customer']=df['Quantity_Ordered'].groupby(df['Order_ID']).transform('sum')
df
```

# WHICH CUSTOMER HAS SPEND MORE MONEY ON OUR PRODUCTS?
